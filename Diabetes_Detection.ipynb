```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.value_counts('Outcome')
```

```python
x = df.drop(columns = 'Outcome')
y = df['Outcome']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
x_test_scaled
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#adding the input and first hidden layer
classifier = Sequential()
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 8))

#adding the second hidden layer
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu'))

#adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = "Adam", loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(x_train,y_train,epochs=100,batch_size=1)
```

```python
y_pred = classifier.predict(x_test_scaled)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm,annot = True)
plt.savefig('h.png')
```

```python
score = classifier.evaluate(x_test_scaled,y_test)
print("Accuracy ",score[1])
```
